```python
import pandas as pd
```

```python
data = pd.read_csv("bronze/Dados_Betfair_Exchange.csv", sep=";")

# Selecinor Time Home
data = data[data["Home"] == "2 de Mayo"]

hthomegoals = data[["Home", "Goals_H_HT", "Goals_A_HT"]]

team_stats = hthomegoals.groupby("Home").agg(
    total_jogos=("Goals_H_HT", "count"),
    total_gols=("Goals_H_HT", "sum"),
    media_gols=("Goals_H_HT", "mean"),
    acima_05=("Goals_H_HT", lambda x: (x > 0.5).sum()),
    acima_15=("Goals_H_HT", lambda x: (x > 1.5).sum()),
    max_gols=("Goals_H_HT", "max"),
    min_gols=("Goals_H_HT", "min")
).reset_index()

team_stats
```
